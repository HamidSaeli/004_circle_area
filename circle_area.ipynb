# Project 004: Calculate the Area of a Circle



```python
# This project helps beginners learn how to calculate the area of a circle using Python.
```

```python
import math  # Import math module to access pi
```

```python
# Step 1: Get input from the user
radius = input("Enter the radius of the circle: ")
```

```python
# Step 2: Convert input to float
radius = float(radius)
```

```python
# Step 3: Calculate the area using the formula: area = Ï€ * r^2
area = math.pi * radius ** 2
```

```python
# Step 4: Print the result
print(f"The area of a circle with radius {radius} is: {area:.2f}")
```
